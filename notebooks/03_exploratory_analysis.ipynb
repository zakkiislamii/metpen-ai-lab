## Apa itu EDA?

**Exploratory Data Analysis (EDA)** = "Bermain" dengan data melalui visualisasi dan ringkasan untuk:

1. **Memahami pola** - distribusi, trend, outlier
2. **Menemukan ide** - pertanyaan apa yang menarik?
3. **Merencanakan analisis** - metode apa yang cocok?
4. **Mengidentifikasi masalah** - ada yang aneh?

**Penting:** EDA bukan untuk menyimpulkan, tapi untuk mengeksplorasi.

## Impor library dan data

```python
# RUN-NOW: Jalankan sel ini apa adanya.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Muat data bersih dari notebook sebelumnya (robust terhadap working directory)
candidate_paths = [
    Path("data/processed/student_performance_cleaned.csv"),
    Path("../data/processed/student_performance_cleaned.csv"),
    Path("/workspaces/metpen-ai-lab/data/processed/student_performance_cleaned.csv"),
]

cleaned_file = next((p for p in candidate_paths if p.exists()), None)
if cleaned_file is None:
    raise FileNotFoundError(
        "File data bersih tidak ditemukan. Jalankan notebook 02 atau periksa folder data/processed."
    )

df = pd.read_csv(cleaned_file)

# Set style untuk visualisasi yang lebih menarik
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12, 5)

print(f"Data dimuat dari: {cleaned_file.resolve()}")
print(f"Data dimuat: {df.shape[0]} baris × {df.shape[1]} kolom")
```

## Chart 1: Histogram - Distribusi final_grade

```python
# RUN-NOW: Jalankan sel ini apa adanya.
plt.figure(figsize=(10, 5))
plt.hist(df['final_grade'], bins=30, color='steelblue', edgecolor='black', alpha=0.7)
plt.xlabel('Nilai Akhir', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.title('Distribusi Nilai Akhir Mahasiswa', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.show()

print("📊 Interpretasi:")
print(f"  Mean (rata-rata): {df['final_grade'].mean():.2f}")
print(f"  Median (tengah): {df['final_grade'].median():.2f}")
print(f"  Std Dev: {df['final_grade'].std():.2f}")
```

### Apa histogram ini menunjukkan?

**Pola yang terlihat:**
- Distribusi nilai akhir mahasiswa
- Apakah distribution normal (bell curve)? Skewed?
- Di mana puncaknya (mode)?
- Ada outlier ekstrim?

**Mengapa histogram cocok:**
- Melihat bentuk distribusi data numerik
- Mengidentifikasi normalitas

**Hati-hati:**
- Banyak bin → terlalu detail, banyak noise
- Sedikit bin → terlalu simplified
- Jumlah histogram tergantung ukuran data dan pertanyaan Anda

## Chart 2: Boxplot - Membandingkan nilai akhir antar program

```python
# RUN-NOW: Jalankan sel ini apa adanya.
plt.figure(figsize=(10, 5))
sns.boxplot(
    data=df,
    x="program",
    y="final_grade",
    hue="program",
    palette="Set2",
    legend=False,
    dodge=False,
    saturation=0.9,
    linewidth=1,
    width=0.7,
    showfliers=True,
    fliersize=4,
    whiskerprops={"linewidth": 1},
    capprops={"linewidth": 1},
    medianprops={"linewidth": 1.5, "color": "#333"},
)
plt.xlabel("Program Studi", fontsize=12)
plt.ylabel("Nilai Akhir", fontsize=12)
plt.title("Perbandingan Nilai Akhir Antar Program", fontsize=14, fontweight="bold")
plt.xticks(rotation=45)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()

# Ringkasan per program
print("📊 Ringkasan per program:")
print(df.groupby("program")["final_grade"].describe())
```

### Apa boxplot ini menunjukkan?

**Elemen boxplot:**
- **Kotak**: Range tengah 50% data (Q1-Q3)
- **Garis di kotak**: Median (Q2)
- **Whisker**: Batas data (±1.5×IQR)
- **Titik**: Outlier (di luar whisker)

**Pola yang dicari:**
- Program mana dengan median tertinggi/terendah?
- Program mana dengan variasi terbesar/terkecil?
- Ada outlier ekstrim di program tertentu?

**Caution:**
- Boxplot tidak menunjukkan distribusi detail
- Bimodal data tidak terlihat jelas di boxplot
- Gunakan dengan histogram untuk gambaran lengkap

## Chart 3: Scatter plot - Hubungan attendance_rate dan final_grade

```python
# RUN-NOW: Jalankan sel ini apa adanya.
plt.figure(figsize=(10, 6))
plt.scatter(df["attendance_rate"], df["final_grade"], alpha=0.5, s=50, color="coral")
plt.xlabel("Tingkat Kehadiran (%)", fontsize=12)
plt.ylabel("Nilai Akhir", fontsize=12)
plt.title("Hubungan Kehadiran dan Nilai Akhir", fontsize=14, fontweight="bold")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Hitung korelasi
corr = df["attendance_rate"].corr(df["final_grade"])
print(f"Korelasi Pearson: {corr:.3f}")
print(
    f"Interpretasi: Hubungan {'positif' if corr > 0 else 'negatif'} "
    f"{'kuat' if abs(corr) > 0.7 else 'sedang' if abs(corr) > 0.4 else 'lemah'}"
)
```

### Apa scatter plot ini menunjukkan?

**Pola yang dicari:**
- Tren positif atau negatif?
- Hubungan linear atau non-linear?
- Ada outlier atau cluster?
- Berapa banyak scatter (variasi)?

**Interpretasi caution:**
- ⚠️ Asosiasi ≠ Kausalitas!
- Mungkin ada confounder (variabel ketiga) yang pengaruhi keduanya
- Strength asosiasi diukur dengan korelasi (kita akan pelajari di notebook 5)

## Chart 4: Bar chart - Distribusi internet access

```python
# RUN-NOW: Jalankan sel ini apa adanya.
internet_counts = df["internet_access"].value_counts()

plt.figure(figsize=(8, 5))
internet_counts.plot(kind="bar", color="lightgreen", edgecolor="black")
plt.xlabel("Tipe Akses Internet", fontsize=12)
plt.ylabel("Jumlah Mahasiswa", fontsize=12)
plt.title("Distribusi Akses Internet", fontsize=14, fontweight="bold")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()

# Tambah persentase
print("Distribusi Akses Internet:")
print((internet_counts / internet_counts.sum() * 100).round(2))
```

### Apa bar chart ini menunjukkan?

**Untuk data kategoris:**
- Frekuensi setiap kategori
- Kategori mana paling/paling sedikit?
- Apakah distribusi balanced atau skewed?

**Interpretasi caution:**
- Bar chart menunjukkan COUNT, bukan rate
- Jika total sampel berbeda, perlu normalisasi ke persentase

## Chart 5: Boxplot tambahan - final_grade by internet access

```python
# RUN-NOW: Jalankan sel ini apa adanya.
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='internet_access', y='final_grade', palette='Set3')
plt.xlabel('Akses Internet', fontsize=12)
plt.ylabel('Nilai Akhir', fontsize=12)
plt.title('Perbandingan Nilai Akhir Berdasarkan Akses Internet', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Apa boxplot ini menunjukkan?

**Variabel yang ditampilkan:**
- Sumbu X: Tipe akses internet mahasiswa (`limited`, `stable`, `unstable`)
- Sumbu Y: Nilai akhir (`final_grade`)

**Mengapa boxplot cocok:**
- Boxplot tepat digunakan untuk membandingkan distribusi satu variabel numerik (`final_grade`) di antara beberapa kelompok kategoris (tipe internet)
- Memperlihatkan median, IQR, dan outlier per kelompok sekaligus dalam satu chart

**Pola yang terlihat:**
- Kelompok `stable` cenderung memiliki median dan distribusi nilai yang sedikit lebih tinggi dibanding `limited` dan `unstable`
- Ketiga kelompok memiliki rentang (IQR) yang hampir serupa, menunjukkan variasi nilai yang tidak jauh berbeda antar kelompok
- Terdapat outlier di sisi bawah pada semua kelompok, menandakan ada sebagian kecil mahasiswa dengan nilai rendah di setiap kategori akses internet

**Hati-hati:**
- ⚠️ Perbedaan yang terlihat di chart ini belum tentu bermakna secara statistik — perlu uji inferensial (misal: ANOVA atau Kruskal-Wallis) untuk mengkonfirmasi
- Akses internet yang lebih baik belum tentu *menyebabkan* nilai lebih tinggi; bisa jadi ada confounder seperti kondisi ekonomi yang memengaruhi keduanya sekaligus
