## Periksa nilai hilang (missing values)

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Hitung missing values
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    "Kolom": missing.index,
    "Jumlah NaN": missing.values,
    "Persentase": missing_percent.values,
})

# Hanya tampilkan kolom dengan nilai hilang
missing_df = missing_df[missing_df["Jumlah NaN"] > 0].sort_values("Jumlah NaN", ascending=False)

if len(missing_df) == 0:
    print("✓ Tidak ada nilai hilang (missing values) dalam dataset")
else:
    print("Kolom dengan nilai hilang:")
    print(missing_df)
```

## Tugas Anda

Sebelum lanjut ke notebook berikutnya, jawab pertanyaan ini:

1. **Kolom apa yang memiliki nilai hilang?** Berapa persentasenya?
2. **Kolom numerik apa yang memiliki range yang tidak realistis?** (Misalnya, nilai negatif ketika harusnya positif)
3. **Ada berapa mahasiswa yang lulus vs tidak lulus?**
4. **Program apa yang memiliki jumlah mahasiswa paling sedikit?**

Tulis jawaban Anda di kertas atau comment di sel ini.

```python
print("JAWABAN TUGAS")
print("=" * 50)

# 1) Kolom dengan nilai hilang
missing_columns = missing_df.copy() if 'missing_df' in globals() else pd.DataFrame()
if len(missing_columns) == 0:
    print("\n1) Kolom dengan nilai hilang: tidak ada.")
else:
    print("\n1) Kolom dengan nilai hilang:")
    for _, row in missing_columns.iterrows():
        print(f"- {row['Kolom']}: {int(row['Jumlah NaN'])} nilai hilang ({row['Persentase']:.2f}%)")

# 2) Kolom numerik yang memiliki range tidak realistis
expected_ranges = {
    "attendance_rate": (0, 100),
    "assignment_score": (0, 100),
    "quiz_average": (0, 100),
    "midterm_score": (0, 100),
    "final_exam_score": (0, 100),
    "lms_logins": (0, None),
    "study_hours_per_week": (0, None),
    "part_time_work_hours": (0, None),
    "commute_minutes": (0, None),
    "previous_gpa": (0, 4.0),
    "final_grade": (0, 100),
    "submission_delay_days": (0, None),
}

unrealistic_columns = []
for column, (lower_bound, upper_bound) in expected_ranges.items():
    if column not in df.columns:
        continue
    series = df[column].dropna()
    invalid_mask = series < lower_bound
    if upper_bound is not None:
        invalid_mask = invalid_mask | (series > upper_bound)
    if invalid_mask.any():
        unrealistic_columns.append(column)

if len(unrealistic_columns) == 0:
    print("\n2) Kolom numerik yang memiliki range tidak realistis: tidak ada.")
else:
    print("\n2) Kolom numerik yang memiliki range tidak realistis:")
    for column in unrealistic_columns:
        print(f"- {column}")

# 3) Distribusi mahasiswa yang lulus vs tidak lulus
print("\n3) Distribusi passed (lulus/tidak):")
passed_counts = df["passed"].value_counts().sort_index()
for label, count in passed_counts.items():
    status = "lulus" if label == 1 else "tidak lulus"
    print(f"- {status}: {count}")

# 4) Program dengan jumlah mahasiswa paling sedikit
program_counts = df["program"].value_counts()
least_program = program_counts.idxmin()
least_count = int(program_counts.min())
print("\n4) Program dengan jumlah mahasiswa paling sedikit:")
print(f"- {least_program}: {least_count} mahasiswa")
```
