## Apa itu statistik deskriptif?

**Statistik deskriptif** = ringkasan numerik yang mendeskripsikan data:

### Ukuran Pusat (Centrality)
- **Mean (rata-rata)**: Penjumlahan semua nilai ÷ jumlah observasi
  - Cocok untuk: data simetris, tanpa outlier ekstrim
  - Sensitif terhadap outlier

- **Median (tengah)**: Nilai di posisi tengah saat diurutkan
  - Cocok untuk: data skewed, ada outlier
  - Robust (tidak terpengaruh outlier)

- **Mode (paling sering)**: Nilai yang paling banyak muncul
  - Cocok untuk: data kategoris atau multimodal

### Ukuran Penyebaran (Dispersion)
- **Std Dev**: Rata-rata deviasi dari mean (dalam unit sama)
- **Variance**: Std Dev kuadrat
- **Range**: Max - Min
- **IQR**: Q3 - Q1 (tengah 50% data)

```python
# RUN-NOW: Jalankan sel ini apa adanya.
import pandas as pd
import numpy as np
from pathlib import Path

# Muat data bersih — dengan fallback untuk berbagai working directory
candidate_paths = [
    Path("data/processed/student_performance_cleaned.csv"),
    Path("../data/processed/student_performance_cleaned.csv"),
    Path("/workspaces/metpen-ai-lab/data/processed/student_performance_cleaned.csv"),
]

cleaned_file = next((p for p in candidate_paths if p.exists()), None)
if cleaned_file is None:
    raise FileNotFoundError(
        "❌ File data bersih tidak ditemukan.\n"
        "Anda perlu menjalankan notebook 02 (02_data_quality_cleaning.ipynb) terlebih dahulu\n"
        "untuk membuat file: data/processed/student_performance_cleaned.csv"
    )

df = pd.read_csv(cleaned_file)

print(f"✓ Data dimuat dari: {cleaned_file.resolve()}")
print(f"✓ Data dimuat: {df.shape[0]} baris × {df.shape[1]} kolom")
```

## Statistik deskriptif variabel numerik utama

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Pilih variabel utama untuk deskriptif
numeric_vars = [
    'attendance_rate',
    'assignment_score', 
    'quiz_average',
    'midterm_score',
    'final_exam_score',
    'study_hours_per_week',
    'final_grade'
]

# Hitung statistik deskriptif
summary = df[numeric_vars].describe().T
summary['median'] = df[numeric_vars].median()
summary['mode'] = df[numeric_vars].apply(lambda x: x.mode()[0] if len(x.mode()) > 0 else np.nan)

print("📊 STATISTIK DESKRIPTIF")
print(summary.round(2))
```

## Interpretasi kolom demi kolom

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Mari fokus pada satu variabel untuk interpretasi detail
var_focus = 'final_grade'

print(f"ANALISIS DETAIL: {var_focus}")
print(f"="*50)
print(f"Count (N):        {df[var_focus].count()}")
print(f"Mean:             {df[var_focus].mean():.2f}")
print(f"Median:           {df[var_focus].median():.2f}")
print(f"Std Dev:          {df[var_focus].std():.2f}")
print(f"Min:              {df[var_focus].min():.2f}")
print(f"Q1 (25%):         {df[var_focus].quantile(0.25):.2f}")
print(f"Q2 (50%/Median):  {df[var_focus].quantile(0.50):.2f}")
print(f"Q3 (75%):         {df[var_focus].quantile(0.75):.2f}")
print(f"Max:              {df[var_focus].max():.2f}")
print(f"IQR (Q3-Q1):      {df[var_focus].quantile(0.75) - df[var_focus].quantile(0.25):.2f}")
print(f"\n→ Interpretasi:")
print(f"   50% mahasiswa mendapat nilai {df[var_focus].quantile(0.50):.1f} atau lebih (median)")
print(f"   Rata-rata {df[var_focus].mean():.1f} dengan deviasi {df[var_focus].std():.1f}")
```

## Kapan gunakan Mean vs Median?

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Bandingkan mean dan median
comparison = pd.DataFrame({
    'Variabel': numeric_vars,
    'Mean': [df[v].mean() for v in numeric_vars],
    'Median': [df[v].median() for v in numeric_vars],
    'Diff (M-Med)': [df[v].mean() - df[v].median() for v in numeric_vars]
})

print("PERBANDINGAN MEAN vs MEDIAN")
print(comparison.round(2))
print("\n→ Interpretasi:")
print("   Jika Mean ≈ Median: Distribusi simetris, gunakan Mean")
print("   Jika Mean >> Median: Ada outlier besar, gunakan Median")
print("   Jika Mean << Median: Ada outlier kecil, gunakan Median")
```

## Statistik deskriptif per grup

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Ringkasan per program
print("STATISTIK DESKRIPTIF FINAL GRADE PER PROGRAM")
print("="*50)

grouped_stats = df.groupby('program')['final_grade'].agg([
    'count',
    'mean',
    'median',
    'std',
    'min',
    'max'
]).round(2)

print(grouped_stats)
```

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Ringkasan per internet access
print("\nSTATISTIK DESKRIPTIF FINAL GRADE PER AKSES INTERNET")
print("="*50)

grouped_stats2 = df.groupby('internet_access')['final_grade'].agg([
    'count',
    'mean',
    'median',
    'std',
    'min',
    'max'
]).round(2)

print(grouped_stats2)
```

## Ringkasan hasil deskriptif

```python
# LEARNER-TASK: Edit sesuai instruksi, lalu jalankan.
# Tulis ringkasan dalam bahasa akademik
program_stats = df.groupby('program')['final_grade'].agg(['mean', 'std'])
internet_stats = df.groupby('internet_access')['final_grade'].mean()

top_program_mean = program_stats['mean'].idxmax()
top_program_std  = program_stats['std'].idxmax()
top_internet     = internet_stats.idxmax()
diff_internet    = internet_stats.max() - internet_stats.min()

summary_text = f"""
RINGKASAN STATISTIK DESKRIPTIF
==============================

NILAI AKHIR:
- Rata-rata: {df['final_grade'].mean():.1f} (SD = {df['final_grade'].std():.1f})
- Median: {df['final_grade'].median():.1f}
- Range: {df['final_grade'].min():.1f} - {df['final_grade'].max():.1f}

PERBANDINGAN ANTAR PROGRAM:
- Program dengan rata-rata tertinggi: {top_program_mean} (mean = {program_stats.loc[top_program_mean, 'mean']:.2f})
- Program dengan variasi terbesar: {top_program_std} (std = {program_stats.loc[top_program_std, 'std']:.2f})

PERBANDINGAN AKSES INTERNET:
- Kelompok dengan nilai tertinggi: {top_internet} (mean = {internet_stats.max():.2f})
- Perbedaan mean antar kelompok: {diff_internet:.2f} poin

CATATAN METODOLOGIS:
- Statistik deskriptif hanya menggambarkan, bukan menjelaskan kausalitas
- Untuk menguji signifikansi perbedaan grup, perlu test inferensial (t-test, ANOVA)
- Untuk mengukur asosiasi variabel, perlu analisis korelasi/regresi
"""

print(summary_text)
```
